## 2. Variables and Data Types
Python is dynamically typed, meaning you don't need to declare the type of variable.

```python
name = "joy"          # String (str)
age = 25              # Integer (int)
height = 1.75         # Floating point (float)
is_learning = True    # Boolean (bool)

print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
print(f"Height: {height}, Type: {type(height)}")
print(f"Is Learning: {is_learning}, Type: {type(is_learning)}")
```

## 3. Arithmetic Operations


```python
a = 10
b = 3

print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Floor Division: {a // b}")
print(f"Exponent (a^b): {a ** b}")
print(f"Modulus (Remainder): {a % b}")
```

## 4. Conditionals
Use `if`, `elif`, and `else` to control the flow of your program.

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

## 5. Loops
`for` loops are great for iterating over sequences, while `while` loops run as long as a condition is true.

```python
print("For Loop Example (0 to 4):")
for i in range(5):
    print(f"Iteration {i}")

print("\nWhile Loop Example:")
count = 3
while count > 0:
    print(f"Countdown: {count}")
    count -= 1
```

## 6. Goal Project: BMI Calculator
Formula: `BMI = weight / (height^2)`

```python
# Example Data
weight_kg = 75
height_m = 1.8

bmi = weight_kg / (height_m ** 2)
print(f"Your BMI is: {bmi:.2f}")

if bmi < 18.5:
    print("Category: Underweight")
elif 18.5 <= bmi < 25:
    print("Category: Normal weight")
elif 25 <= bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
```
